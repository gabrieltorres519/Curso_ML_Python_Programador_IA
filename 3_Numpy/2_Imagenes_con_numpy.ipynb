# Manejando im√°genes con Numpy

```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
# Cargando la imagen a un Numpy array

array_carro = plt.imread('car_c.jpg')
array_carro.shape # Filas, Columnas, Capas
```

```python
# Visualizamos la imagen

plt.imshow(array_carro, interpolation="nearest") # Interpolaci√≥n nearest para que pueda aplicar los colores correctos y visualizarlos
```

```python
# Extrayendo una de las capas de la im√°gen

capa_rojo = array_carro[:,:,0]
plt.imshow(capa_rojo, cmap=plt.cm.Reds_r) # Tenemos que indicar que lo que se env√≠a es una capa de color, para que los valores o intensidades
                                          # se interpreten como intensidades de rojo
```

```python
# Extrayendo una de las capas de la im√°gen

capa_verde = array_carro[:,:,1]
plt.imshow(capa_verde, cmap=plt.cm.Greens_r) # Tenemos que indicar que lo que se env√≠a es una capa de color, para que los valores o intensidades
                                          # se interpreten como intensidades de verde
```

```python
# Extrayendo una de las capas de la im√°gen

capa_azul = array_carro[:,:,1]
plt.imshow(capa_azul, cmap=plt.cm.Blues_r) # Tenemos que indicar que lo que se env√≠a es una capa de color, para que los valores o intensidades
                                          # se interpreten como intensidades de azul
```
