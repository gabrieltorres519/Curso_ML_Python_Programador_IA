```python
import pandas as pd
```

```python
dataset = pd.read_csv("titanic.csv")
```

```python
# Acceder a la cantidad de filas que queramos 

dataset.head(10)
```

### Uso de loc e iloc

```python
# Creamos un objeto Series

animales = pd.Series(["Elefantes","Leopardo","Panda","Cocodrilo","Serpiente"],
                     index=[0,3,9,8,3])

animales
```

```python
# Seleccionar todos los elementos con un mismo index específico con loc

animales.loc[3] 
```

```python
# Para localizar por posición usamos iloc
animales.iloc[3]
```

```python
# En el contexto de un DataFrame loc se usa para extraer los datos de una fila

dataset.loc[3]
```

```python
# En el contexto de un DataFrame iloc se usa para extraer un grupo de filas 
# por ejemplo las 3 primeras filas 

dataset.iloc[0:3] 
```

```python
# Extraer los datos de una columna, es la manera más utilizada debido a que los nombres de las columnas 
# a veces llevan espacios  
dataset['Name']
```

```python
# Otra manera de extraer la columna que solo funciona si el dataset está formado con Pandas  

dataset.Name
```

```python
# Se pueden también crear filtros como si se tratara de una consulta SQL

dataset[dataset["Age"] < 2]
```

```python
# Se pueden realizar operaciones

dataset["Age"].mean()
```
